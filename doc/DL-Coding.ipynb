## Performance considerations to specifying int64 vs. int?
[18:33]? 
Here the key performance issue was to use the smallest number of bits that could represent the column. 
If we used `int8` for `item_nbr`, the maximum item_nbr is bigger than 255 and it will not fit. 
If we used int64 for the `store_nbr`, it is using more bits than necessary. 
To avoid running out of RAM, we do not want to use more memory than necessary. 
When working with large datasets, reading and writing to RAM is slower than CPU operations. 
As a rule of thumb, smaller data types will run faster if we use Single Instruction Multiple Data (SIMD) vectorized code, it can pack more numbers into a single vector to run at once.

# RAM problems and `proc_all_mp()`
If we have RAM problems, maybe because of this:
`proc_all_mp()` is processing every sentence across multiple processes [25:59]: 
We find how many CPUs are available, divide it by two (because normally with hyper-threading they don’t actually all work in parallel), then in parallel run `proc_all`. 
It creates separate Python processes for every CPU. 
If there are many cores, then many Python processes loading data...potentially use up all RAM. 
Alternative is to replace proc_all_mp by just proc_all to use less RAM. 
Or use less cores. 
At the moment, we are calling partition_by_cores which calls partition on a list and asks to split it into a number of equal length things according to how many CPUs you have. 
So you could replace that to split into a smaller list and run it on less things.

## * the "splat" operator
"*" is the "splat" operator: It takes a list as input, and expands it into actual positional arguments in the function call.

So if uniqueCrossTabs was [ [ 1, 2 ], [ 3, 4 ] ], then itertools.chain(*uniqueCrossTabs) is the same as saying itertools.chain([ 1, 2 ], [ 3, 4 ])

This is obviously different from passing in just uniqueCrossTabs. In your case, you have a list of lists that you wish to flatten; what itertools.chain() does is return an iterator over the concatenation of all the positional arguments you pass to it, where each positional argument is iterable in its own right.

In other words, you want to pass each list in uniqueCrossTabs as an argument to chain(), which will chain them together, but you don't have the lists in separate variables, so you use the * operator to expand the list of lists into several list arguments.

## inplace

### Q: Why is inplace=True in the LeakyReLU L12?
[17:54]? Thanks for asking! 
This is a really important memory technique. 
The conv_layer is the lowest level thing.
So everything in our ResNet (once it’s all together),
is going to be many conv_layer’s. 
Without `inplace=True`, it’s going to create a separate piece of memory for the output of the ReLU, which is unnecessary.

Another example is that the original forward in ResLayer looked like:
```
def forward(self, x): return x + self.conv2(self.conv1(x))
```
Recall in PyTorch every function has an underscore (_) suffix version,
which tells it to do it in-place. 
 + is equivalent to add and in-place version of add is add_ 
so this will reduce memory usage:

```
def forward(self, x): return x.add_(self.conv2(self.conv1(x)))
```
If you forget the `inplace=True` you may end up with a larger memory use. 
You can also do that with dropout if you have dropout. 
Here are what to look out for:
- Dropout
- All the activation functions
- Any arithmetic operation

# Hardware topics
## random number generators 
on computers aren't random.
Given some initial starting point, e.g. 42, a pseudo-random number
generator is a mathematical function that generates a deterministic, 
always the same sequence of numbers such that those numbers are designed to be as
uncorrelated with the previous number as possible, and as unpredictable as possible.<br>
For real random numbers the way to do it is to buy a 
[hardware random number generator](https://en.wikipedia.org/wiki/Hardware_random_number_generator).


## Dont use clusters of computers
When Bremen created RFs (1999) we always wanted to use all the data we had.
Nowadays we have too much data.
Many people tend to fire up a spark cluster on many machines, which makes no sense.
If they just use a subsample each time it can be done on one machine.
The I/O overhead of spark is huge, so on a single machine it can be faster.
Also it is easier to write the algorithms, use SKlearn, easier to visualize, cheaper, etc.
JH avoided distributed computing, even 25 years ago, 
whatever we could do with a cluster now we could do with a single machine in 5 years.
Focus on being as good as possible with a single machine, more interactive and more iterative.
