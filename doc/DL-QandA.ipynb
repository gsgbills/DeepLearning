# Training 

## Q: Can we unfreeze just specific layers? 
[51:03] 
Not now. `learn.unfreeze_to(n)` will unfreeze layers from layer n onwards. 
Not helpful, probably because of **Discriminative Learning Rates (DLR)**, 
and the optimizer can learn just as much as it needs to. 
It is helpful when a big memory intensive model is running out of GPU:
the fewer layers unfreezed, the less memory and time it takes.

## Q: What is the difference between precompute=True and unfreeze?
- We started with a pre-trained network (arch)
- We added a couple of layers at the end which started out with random initializations. 
- With everything frozen and precompute=True, we are only learning the layers that we added.
- With `precompute=True`, data augmentation does not do anything, because we are showing exactly the same activations each time.
- When we set `precompute=False` data augmentation works, it is recalculating all of the activations from scratch.   (As it is still frozen, we are only training the layers we added).
- Finally, we `unfreeze`, i.e., “now go ahead and change all of these earlier convolutional filters”.

## Q: Why not just set precompute=False from the beginning? 
The only reason to have `precompute=True` is it is much faster (10 or more times). 
If working with a large dataset, it can save time, but there is no accuracy reason ever to use precompute=True.


## Q: Does reducing the batch size only affect the speed of training? 
[1:43:34] 
If you are showing it less images each time, then it is calculating the gradient with less images — hence less accurate. 
In other words, knowing which direction to go and how far to go in that direction is less accurate. 
As the batch size is smaller it is more volatile. 
It impacts the optimal learning rate that you would need to use.
But in practice, dividing the batch size by 2 vs. 4 does not seem to change things very much. 
If you change the batch size by much, you should re-run the learning rate finder to check.

# Weight Space

## Q: Why do smoother surfaces correlate to more generalized networks? 
[55:28]



Say you have something spiky (blue line). 
X-axis shows how good the model is at recognizing (eg dogs vs. cats) as you change a particular parameter. **Generalizable** means that it works well with a (slightly?) different dataset. 
A slightly different dataset may have a (slightly) different relationship between this parameter and how cat-like vs. dog-like it is, i.e., it may look like the red line. 

If we end up at the blue pointy part, it will not do a good job on this slightly different dataset. 
Else, if we end up on the wider blue part, it will still do a good job on the red dataset.

See [Do smoother areas of the error surface lead to better generalization?](https://medium.com/@radekosmulski/do-smoother-areas-of-the-error-surface-lead-to-better-generalization-b5f93b9edf5b) for a relevant interesting discussion. 

## Q: Why do we care about the resilency of where we are in the weight space?
Isn’t low loss the only thing that we care about? Low loss == being close to global minimum.<br>
A: We only care about loss in the validation/test sets. 
But our training set can’t see that data. 
So we need to know whether slightly different datasets might have very different losses.<br>

Also, it is interesting to consider this is as a property of various training algorithms:
- to what extent they not only produce a low train error, but also 
- what measures they take that can help us converge at a nice solution
I wonder if we could actually measure how ‘spiky’ the surface at the solution is, 
and compare different training methods and architectures to see which finds the smoothest surface.

Useful idea: We can tell something about our ability to generalize based on the shape of the surrounding error surface on the train set. 
For the cyclical learning rates it is also a function of how many potential areas we can evaluate in a given timeframe - many more so than we would be able to without the annealing!

See [Why do we care about resilency of where we are in the weight space?](http://forums.fast.ai/t/why-do-we-care-about-resilency-of-where-we-are-in-the-weight-space/7323)

## Q: How do we deal with unbalanced dataset? 
[01:38:46] 
The dogs-breeds dataset is not balanced (60-100 per category) but it is not unbalanced enough 
to worry. 
A recent paper says the best way to deal with very unbalanced dataset is to make copies of the rare cases.
From [A systematic study of the class imbalance problem in convolutional neural networks](https://arxiv.org/abs/1710.05381) :
- the effect of class imbalance on classification performance is detrimental; 
- the method of addressing class imbalance that emerged as dominant in almost all analyzed scenarios was oversampling; 
- oversampling should be applied to the level that totally eliminates the imbalance, whereas undersampling can perform better when the imbalance is only removed to some extent; 
- as opposed to some classical machine learning models, oversampling does not necessarily cause overfitting of CNNs; 
- thresholding should be applied to compensate for prior class probabilities when overall number of properly classified cases is of interest

## Q: Why not border or padding to make it square? 
Typically JH does not do much padding, but instead a little bit of zooming. 
*Reflection padding* works well with satellite imagery. 
Generally speaking, using TTA plus data augmentation, as large image as possible is best. 
Also, having fixed crop locations plus random contrast, brightness, rotation changes might be better for TTA.

## Q: How many images should we use as a validation set? 
[01:26:28] 
Using 20% is fine unless the dataset is small — then 20% is not enough. 
If we train the same model multiple times and get very different validation set results, 
then the validation set is too small. <br>
If the validation set is smaller than 1000, it is hard to interpret how well we are doing. 
If we care about the third decimal place of accuracy, but we only have 1000 items in the validation set, 
a single image changes the accuracy. 
If you care about the difference between 0.01 and 0.02, you want that to represent 10 or 20 rows. 
Normally 20% seems to work fine.

# Qs on Discriminative  learning rates (DLRs)

## Question: Why don’t we start training with DLRs rather than training the last layers alone? 
[01:50:30]
You can skip training just the last layer and go straight to DLR, but you probably do not want to. Convolutional layers contain pre-trained weights, they are not random, e.g., 
for things that are close to ImageNet, they are really good; 
for things that are not close to ImageNet, they are still better than nothing. 
All of our fully connected layers, however, are totally random. 
Therefore, we always want to make the fully connected weights better than random by training them a bit first. 
Otherwise if we go straight to unfreeze, we would be fiddling 
with those early layer weights when the later ones are still random... — probably not what we want.

## Question: When you use DLRs, do the 3 learning rates spread "evenly" across the layers? 
[01:55:35] 
The fast.ai library has a concept of “layer groups”. 
DNNs, e.g., ResNet50, have 100s of layers, and we do not want to write 100s of learning rates.
The library decided how to split them.
The last number always refers to just the Fully Connected (FC) layers that we randomly initialized and added.

## Question: Learning rate finder for a small dataset returned strange number and the plot was empty 
[01:58:57] — 
The learning rate finder will go through a mini-batch at a time. 
For a tiny dataset there are not enough mini-batches. 
The trick is to make the batch size very small like 4 or 8.

# Categorical vs Continous Variables

## Q: Do you ever bin continuous variables?
[31:02] 
No. But we could, eg for max temperature, group into 0–10, 10–20, 20–30, and call that categorical. 
A (4Q17?) paper found that sometimes binning can be helpful.

## Q: If year is a category, what happens when a model encounters a year it has never seen? 
[31:47] 
It will be treated as an unknown category. 
Pandas has a special category called `unknown` and if it sees a category it has not seen before, 
it gets treated as unknown.

## Q: How does using dates and times as categorical affects seasonality? 
[01:06:59] 
`add_datepart` takes a data frame and a column name. 
It optionally removes the column from the data frame,
and replaces it with lots of columns representing all of the useful information about that date.
These include day of week, day of month, month of year, etc (basically everything Pandas gives us).

## Question: Are we using any time series in any of these?
[01:15:01] Indirectly, yes. 
As we just saw, we have a day of week, month of year, etc in our columns.
Most of them are being treated as categories.
We are building a distributed representation of January, Sunday, and so on. 
We are not using any classic time series techniques.
All we are doing is true fully connected layers in a DNN. 
**NB: The embedding matrix is able to deal with things like day of week periodicity 
in a much richer way than than any standard time series techniques.**

## Q: What is the advantage of using embedding matrices over one-hot-encoding? 
[59:23] 
For the day of week example above, instead of the 4 numbers, we could have easily passed 7 numbers (e.g. [0, 1, 0, 0, 0, 0, 0] for Sunday). 
That also is a list of floats and that would work.
That is how categorical variables have been used in statistics (called “dummy variables”). 
The problem is, the concept of Sunday could only be associated with a **single** floating-point number. 
So it gets just a single "dimension". 
With embeddings, Sunday is a concept in 4 dimensional space. 
These embedding vectors capture rich semantic concepts. 
For example, if weekends have a different behavior, you see that 
Saturday and Sunday will have some particular number higher.

## Q: Are embeddings suitable for certain types of variables? 
[01:02:45] 
Embeddings are suitable for any categorical variables. 
It cannot work well for something with too high cardinality. 
If you had 600K rows and a variable had 600K levels, that is just not a useful categorical variable. 
The Rossman 3rd place decided that everything that was not too high cardinality, 
they put as categorical. 
**NB: If we can make a variable categorical, we should**, so it can learn a rich distributed representation; 
A continuous variable, the most it can do is to try and find a single functional form that fits it well.

## Question: How do we choose the size of bptt? [21:36]
There are a couple things to think about:

- mini-batch matrix has a size of bs (# of chunks) by bptt so your GPU RAM must be able to fit that by your embedding matrix. So if you get CUDA out of memory error, you need reduce one of these.
- If your training is unstable (e.g. your loss is shooting off to NaN suddenly), then you could try decreasing your bptt because you have less layers to gradient explode through.
- If it is too slow [22:44], try decreasing your bptt because it will do one of those steps at a time. for loop cannot be parallelized (for the current version). 
There is a recent thing called QRNN (Quasi-Recurrent Neural Network) which does parallelize it and we hope to cover in part 2.
- So pick the highest number that satisfies all these.
